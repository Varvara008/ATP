```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread


image = imread("images/a1.jpg")
image
```

```python
image.shape
```

Переделываем в черно-белую картинку

```python
grey_image = np.mean(image, axis=2)
grey_image
```

```python
plt.imshow(grey_image, cmap="gray")
plt.show()
```

```python
cropped = grey_image[400:800, 600:900]
plt.imshow(cropped, cmap="gray")
plt.show()
```

```python
from skimage.transform import resize


resized = resize(grey_image, (640, 480))
plt.imshow(resized, cmap="gray")
plt.show()
```
