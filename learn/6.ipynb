```python
import keras as k
import numpy as np

#    (10x10)
img = np.random.rand(1, 100, 100, 1)

#     
model = k.models.Sequential([
    k.layers.Conv2D(1, (3, 3), input_shape=(100, 100, 1), activation='relu')
])

#  
feature_map = model.predict(img)

# 
plt.imshow(feature_map[0, :, :, 0], cmap='gray')
plt.show()
```

```python
import keras as k


model = k.models.Sequential([
    k.layers.Conv2D(32, (3, 3), activation="relu"),
    k.layers.MaxPooling2D(),
    k.layers.Flatten(),
    k.layers.Dense(128, activation="relu"),
    k.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss='binary_crossentropy', metrics=['accuracy'])
```
